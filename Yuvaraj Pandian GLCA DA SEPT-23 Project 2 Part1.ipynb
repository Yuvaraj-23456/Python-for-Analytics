Q6) Check if there are missing values present in the column Rating, If any? drop them and and create a new column  'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python
data.shape[0]
```

```python
data.Rating.isna().sum()
```

```python
blank=data[data["Rating"].isna()].index
```

```python
blank
```

```python
data.drop(blank, inplace=True)
```

```python
data.shape[0]
```

```python
data.shape[1]
```

```python
def Rating_category(value):
    if value <= 3.5:
        return "Low"
    elif value > 3.5:
        return "High"
```

```python
data["Rating_category"] = data['Rating'].map(Rating_category)
```
