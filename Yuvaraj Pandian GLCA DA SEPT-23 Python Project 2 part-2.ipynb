Q3) Display the percentage distribution of the data in each year using the pie chart.

```python
Each_year = data.groupby('year').size()/len(data)*100
plt.figure(figsize=(9,9))
plt.pie(Each_year, labels=Each_year.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year')
plt.show()
```

Q8) Display the correlation values using a plot and add your inferences.

```python
data_subset = data[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]
corr_matrix = data_subset.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix. Heat Map")
plt.show()
```

```python
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            corr_value = corr_matrix.loc[col1, col2]
            if corr_value > 0.7:
                print(f"There is a strong positive correlation between {col1} and {col2}.")
            elif corr_value < -0.7:
                print(f"There is a strong negative correlation between {col1} and {col2}.")
            elif 0.3 < corr_value < 0.7:
                print(f"There is a moderate positive correlation between {col1} and {col2}.")
            elif -0.7 < corr_value < -0.3:
                print(f"There is a moderate negative correlation between {col1} and {col2}.")
            else:
                print(f"There is a weak or no correlation between {col1} and {col2}.")
```
