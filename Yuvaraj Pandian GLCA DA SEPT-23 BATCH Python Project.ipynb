Q11.Find the title of the video which has the maximum number of likes and the title of the video having
minimum likes and write your inferences.

```python
max_likes = data["likes"].max()
max_likes_title = data[data["likes"] == max_likes]['title'].values[0]
print("The title of the video with maximum likes is:")
print(max_likes_title)
```

```python
min_likes = data["likes"].min()
min_likes_title = data[data["likes"] == min_likes]["title"].values[0]
print("The title of the video with minimum likes is:")
print(min_likes_title)
```

The most liked video is the BTS () \'Dynamite\' Official MV having more likes and the minimum is of Kim Kardashian\'s Must-See Moments on "Saturday Night Live"  likes. It is also observed that the videos having higher number of likes are having higher number of views.

Q13.Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.

```python
corr = data["view_count"].corr(data["dislikes"])
corr
```

```python
import matplotlib.pyplot as plot
plot.scatter(data["view_count"], data["dislikes"])
plot.xlabel("Views")
plot.ylabel("Dislikes")
plot.title("Correlation between Views and Dislikes")
plot.show()
```

The correlation between views and dislikes is 0.684. It means it's having a direct relation and increase in one leads to increase in another and vice-versa.
