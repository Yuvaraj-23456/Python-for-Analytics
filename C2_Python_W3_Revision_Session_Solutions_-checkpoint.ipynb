<font size=3>Q. Merge the given two dataframes on "Employee_ID"

```python
pd.merge(my_df1, my_df2, on = ['Employee_Id'], how = 'inner')                     # inner means intersection
```

```python
pd.merge(my_df1, my_df2, on = ['Employee_Id'], how = 'outer')                      # outer means union (include everything)
```

```python
pd.merge(my_df1, my_df2, on = ['Employee_Id'], how = 'left')              # LEFT gives importance to left dataframe(my_df1)
```

```python
pd.merge(my_df1, my_df2, on = ['Employee_Id'], how = 'right')             # right gives importance to right dataframe(my_df2)
```

##### Lets Create a multiindex Dataframe

`MultiIndex.from_product`
- We can create MultiIndex Dataframes using the from_product function in Pandas.
- This function takes an iterable of iterables, and creates a MultiIndex Dataframe from it. 

```python
# Row level multiindex Dataframe
iterables = [['Tony', 'Rickie', 'Chap', 'Blake'], ["Excel", "Python", "SQl"]] 
index = pd.MultiIndex.from_product(iterables, names=["Names", "Course"]) 
data = [ [98, 96], [96, 97], [85, 89], [92, 95], [99, 94], [87, 93], [91, 89], [90, 93], [82, 87], [97, 90], [95, 91], [84, 91]]

output_df = pd.DataFrame(data, index=index, columns=["Exam 1", "Exam 2"]) 
output_df
```

```python
## Column level mutliindex dataframe

iterables = [['Exam Paper 1', 'Exam paper 2'], ['Marks']]
index = pd.MultiIndex.from_product(iterables) 
data = [[98, 96], [96, 97], [85, 89], [92, 95], [99, 94], [87, 93], [91, 89]]

output_df1 = pd.DataFrame(data, index=['Tony', 'Rickie', 'Chap', 'Blake',"Excel", "Python", "SQl"], columns=index ) 
output_df1
```

<font color='Green' size=4> **Data and Datatypes** <font> 

**Data**
- A very common kind of data we deal with is Structured Data i.e, data in tabular format. We have rows and columns- rows represent individual observations columns represent attributes/features/variables of the observations.
- Besides the regular tabular data, we also encounter Unstructured data - Text, images, audio, video, analog data etc
- Despite the kind of data, we always convert them into arrays of numbers.

**Datatypes**
- Since Python is an object oriented language we have a wide range of specialized datatypes.
- Some common built-in data types are list, dict, str, int, tuple, set etc

<font color='Green' size=4> **Variables** <font> 

**Qualitative Variable**
- A broad category for any variable that can't be counted (i.e, has no numerical value)
- Nominal and ordinal variables fall under this category.

**Quantitative Variable**
- A broad category for any variable that can be counted, or has a numerical value associated with it.
- Examples of variables that fall into this category include discrete and ratio variables.

<font color='Green' size=4> **Descriptive data measures** <font> 

**Measures of central tendency**
- mean
- median
- mode

**Measures of dispersion**

Central tendency describes the central point in a data set whereas dispersion or variability describes the spread of the data.

`Variance`: Quantity expressing by how much each datapoint differs from the average(mean).

`standard deviation`: Square root of variance.

`range`: The difference between the highest and lowest value of a set

`IQR`: Range computed only for the middle 50% of the data

`skewness` (shape of the data)

<font color='Green' size=4> **Covariance and Correlation** <font> 

**Covariance** 

To measure how one variable varies with respect to the variation of another variable, we compute Covariance.
- The metric only tells us to what extent two variables change together.
- The metric cannot be the only evidence to conclude that variation in one variable is causing the variation in the other variable.
- A positive value indicates that the two variables move in the same direction.
- A negative value indicates that the two variables move in the opposite direction to each other.


**Correlation**

We can only determine the direction of the relationship between two variables but to determine the strength of the relationship, we compute the Correlation coefficient of the two variables.
- Unlike covariance, correlation coefficient is independant of scale
- The range of values correlation coefficient can take is from -1 to +1
- A value closer to -1 or +1 means a strong correlation and a value close to zero means weak correlation
